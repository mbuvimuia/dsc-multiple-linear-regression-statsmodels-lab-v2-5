## The Ames Housing Dataset

The [Ames Housing dataset](http://jse.amstat.org/v19n3/decock.pdf) is a newer (2011) replacement for the classic Boston Housing dataset. Each record represents a residential property sale in Ames, Iowa. It contains many different potential predictors and the target variable is `SalePrice`.

````python
import pandas as pd
ames = pd.read_csv("ames.csv", index_col=0)
ames
````

````python
ames.describe()
````

We will focus specifically on a subset of the overall dataset. These features are:

```
LotArea: Lot size in square feet

1stFlrSF: First Floor square feet

GrLivArea: Above grade (ground) living area square feet
```

````python
ames_subset = ames[['LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']].copy()
ames_subset
````

## Step 1: Visualize Relationships Between Features and Target

For each feature in the subset, create a scatter plot that shows the feature on the x-axis and `SalePrice` on the y-axis.

````python
# Your code here - import relevant library, create scatter plots
import matplotlib.pyplot as plt
import seaborn as sns

fig , ax = plt.subplots(1, 3, figsize=(15, 5))

sns.scatterplot(x='LotArea', y='SalePrice', data=ames_subset, ax=ax[0])
sns.scatterplot(x='1stFlrSF', y='SalePrice', data=ames_subset, ax=ax[1])
sns.scatterplot(x='GrLivArea', y='SalePrice', data=ames_subset, ax=ax[2])
````

````python
# Your written answer here - do these seem like good candidates for linear regression?
# Yes, all three variables seem to have a linear relationship with SalePrice.
# The relationship between LotArea and SalePrice is not very strong.
````

## Step 2: Build a Simple Linear Regression Model

Set the dependent variable (`y`) to be the `SalePrice`, then choose one of the features shown in the subset above to be the baseline independent variable (`X`).

Build a linear regression using StatsModels, describe the overall model performance, and interpret its coefficients.

````python
# Your code here - define y and baseline X
y = ames_subset['SalePrice']
X = ames_subset['GrLivArea']
````

````python
# Your code here - import StatsModels, fit baseline model, display results
import statsmodels.api as sm

simple_model = sm.OLS(y, sm.add_constant(X))
results = simple_model.fit()
results.summary()
````

````python
# Your written answer here - interpret model results
'''
The model is statistically significant overall with an F-Statistic p-value well below 0.05
The model explains about 50% of the variance in SalesPrice
The models coefficients(GrLivArea and const) are both statistically significant with t-statistics p-values well below 0.05
If GrLiveArea is 0, we would expect SalesPrice to be 18569.0259
For each increase of 1 square foot of above ground living area, we would expect SalesPrice to increase by about 107 shillings
'''
````

## Step 3: Build a Multiple Linear Regression Model

For this model, use **all of** the features in `ames_subset`.

````python
# Your code here - define X
X_all = ames_subset[['LotArea', '1stFlrSF', 'GrLivArea']]
y = ames_subset['SalePrice'] 
````

````python
# Your code here - fit model and display results
multiple_model = sm.OLS(y, sm.add_constant(X_all))
results = multiple_model.fit()
results.summary()
````

````python
# Your written answer here - interpret model results. Does this model seem better than the previous one?
'''
The model is statistically significant overall with an F-statistic p-value well below 0.05
The model explains about 56% of the variance in SalesPrice. This is a slight improvement over the simple model with a 6% increase.
The model coefficients for GrLivArea, const and 1stFlrSF are all statistically significant with t-statistics p-values well below 0.05. 
Except for LotArea, which has a p-value of 0.051 which is slightly above 0.05.
For each increase of 1 square foot of above ground living area, we would expect SalesPrice to increase by about 81 shillings
For each increase of 1 square foot of first floor living area, we would expect SalesPrice to increase by about 60 shillings
For each increase of 1 square foot of lot area, we would expect SalesPrice to increase by about 0 shillings.


The model seems to be better than the previous one as it explains more variance in SalesPrice. 
However, LotArea does not seem to be a significant predictor of SalesPrice in this model. Thus we might consider removing it from the model.
'''


````

## Step 4: Create Partial Regression Plots for Features

Using your model from Step 3, visualize each of the features using partial regression plots.

````python
# Your code here - create partial regression plots for each predictor
fig = plt.figure(figsize=(12,8))
sm.graphics.plot_partregress_grid(results, fig=fig)
plt.show();
````

````python
# Your written answer here - explain what you see, and how this relates
# to what you saw in Step 1. What do you notice?

'''
The partial regression plots show the relationship between each predictor and the target variable, SalesPrice.
The relationship between GrLivArea and SalesPrice is still linear, as seen in the scatter plot in Step 1.
The relationship between 1stFlrSF and SalesPrice is also linear, as seen in the scatter plot in Step 1.
The relationship between LotArea and SalesPrice is not linear, as seen in the scatter plot in Step 1.
This confirms our earlier observation that LotArea might not be a good predictor of SalesPrice.
'''
````

## Level Up (Optional)

Re-create this model in scikit-learn, and check if you get the same R-Squared and coefficients.

````python
# Your code here - import linear regression from scikit-learn and create and fit model
from sklearn.linear_model import LinearRegression
sklearn_multiple_model = LinearRegression()
# passing in the same X and y, although the order is reversed
sklearn_multiple_model.fit(X_all, y)
````

````python
# Your code here - compare R-Squared
print(f"""
StatsModels R-Squared:    {results.rsquared}
scikit-learn R-Squared:   {sklearn_multiple_model.score(X_all, y)}
""")
````

````python
# Your code here - compare intercept and coefficients
print(f"""
StatsModels intercept:     {results.params["const"]}
scikit-learn intercept:    {sklearn_multiple_model.intercept_}

StatsModels coefficients:  {results.params[1:].values}
scikit-learn coefficients: {sklearn_multiple_model.coef_}
""")
````
